```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('../data2/train_1205.txt', sep='\t')
train.head()
```

```python
test = pd.read_csv('../data2/test_1205.txt', sep='\t')
test.head()
```

```python
def parse_str(d):
    return np.array(list(map(float, d.split())))
```

```python
def parse_list_1(d):
    if d == '-1':
        return [0]
    return list(map(lambda x: int(x[1:]), str(d).split(',')))
```

```python
train.columns.values.tolist()
```

```python
test.columns.values.tolist()
```

```python
def parse_map(d):
    if d == '-1':
        return {}
    return dict([int(z.split(':')[0][1:]), float(z.split(':')[1])] for z in d.split(','))
```

```python
train['关注话题'] = train['关注话题'].apply(parse_list_1)
train['问题标题切词编码'] = train['问题标题切词编码'].apply(parse_list_1)
train['问题描述切词编码'] = train['问题描述切词编码'].apply(parse_list_1)
train['问题绑定话题'] = train['问题绑定话题'].apply(parse_list_1)
train['感兴趣话题'] = train['感兴趣话题'].apply(parse_map)
train.head()
```

```python
test['关注话题'] = test['关注话题'].apply(parse_list_1)
test['问题标题切词编码'] = test['问题标题切词编码'].apply(parse_list_1)
test['问题描述切词编码'] = test['问题描述切词编码'].apply(parse_list_1)
test['问题绑定话题'] = test['问题绑定话题'].apply(parse_list_1)
test['感兴趣话题'] = test['感兴趣话题'].apply(parse_map)
test.head()
```

```python
topicmap = pd.read_csv('../data/topic_vectors_64d.txt', 
                          names=['id', 'embed'], sep='\t')
topicmap['embed'] = topicmap['embed'].apply(parse_str)
topicmap['id'] = topicmap['id'].apply(lambda x: int(x[1:]))
topicmap.head()
```

```python
topicmap.shape
```

```python
topic_vector_dict = dict(zip(np.array(topicmap['id']), np.array(topicmap['embed'])))
```

```python
import pickle
from tqdm import tqdm, tqdm_notebook, _tqdm_notebook, tqdm_pandas
```

```python
def topic2v(x):
    try:
        tmp = topic_vector_dict[x[0]]
    except:
        tmp = np.zeros(64)
    for i in x[1:]:
        tmp = tmp + topic_vector_dict[i]
    if len(tmp) == 0:
        return np.zeros(64)
    return (tmp / len(x))
```

```python
tqdm.pandas(desc="topic2v...")
train['focus_topics']=train['关注话题'].progress_apply(lambda x:topic2v(x))
print('finished!')
```

```python
tqdm.pandas(desc="topic2v...")
test['focus_topics']=test['关注话题'].progress_apply(lambda x:topic2v(x))
print('finished!')
```

```python
tqdm.pandas(desc="topic2v...")
train['question_topics']=train['问题绑定话题'].progress_apply(lambda x:topic2v(x))
print('finished!')
```

```python
tqdm.pandas(desc="topic2v...")
test['question_topics']=test['问题绑定话题'].progress_apply(lambda x:topic2v(x))
print('finished!')
```

```python
wordmap = pd.read_csv('../data/word_vectors_64d.txt', 
                          names=['id', 'embed'], sep='\t')
wordmap['embed'] = wordmap['embed'].apply(parse_str)
wordmap['id'] = wordmap['id'].apply(lambda x: int(x[1:]))
wordmap.head()
```

```python
wordmap.shape
word_vector_dict = dict(zip(np.array(wordmap['id']), np.array(wordmap['embed'])))
```

```python
def word2v(x):
    try:
        tmp = word_vector_dict[x[0]]
    except:
        tmp = np.zeros(64)
    for i in x[1:]:
        tmp = tmp + word_vector_dict[i]
    if len(tmp) == 0:
        return np.zeros(64)
    return (tmp / len(x))
```

```python
tqdm.pandas(desc="word2v...")
train['title_questions']=train['问题标题切词编码'].progress_apply(lambda x:word2v(x))
print('finished!')
```

```python
tqdm.pandas(desc="word2v...")
test['title_questions']=test['问题标题切词编码'].progress_apply(lambda x:word2v(x))
print('finished!')
```

```python
tqdm.pandas(desc="word2v...")
test['describe_questions']=test['问题描述切词编码'].progress_apply(lambda x:word2v(x))
print('finished!')
```
